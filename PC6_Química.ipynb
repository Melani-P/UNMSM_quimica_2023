# Gráfica a

image.png

```python
df_a = pd.read_csv('data_a.csv')
df_a
```

Medidas de tendencia central, siendo a, b y c tres temperaturas diferentes y a < b < c

```python
print('La media de la velocidad molecular en a es:', df_a[df_a['z'] == 'a']['x'].mean())
print('La media de la velocidad molecular en b es:', df_a[df_a['z'] == 'b']['x'].mean())
print('La media de la velocidad molecular en c es:', df_a[df_a['z'] == 'c']['x'].mean())
```

```python
print('La mediana de la velocidad molecular en a es:', df_a[df_a['z'] == 'a']['x'].median())
print('La mediana de la velocidad molecular en b es:', df_a[df_a['z'] == 'b']['x'].median())
print('La mediana de la velocidad molecular en c es:', df_a[df_a['z'] == 'c']['x'].median())
```

Tanto la medía como la mediana en el eje "x" de las tres temperaturas es la misma debido a que en el dataframe se completan datos con 0, sin embargo en la gráfica se puede observar que a mayor temperatura la velocidad molecular aumenta.

```python
print('La media del número de moléculas en a es:', df_a[df_a['z'] == 'a']['y'].mean())
print('La media del número de moléculas en b es:', df_a[df_a['z'] == 'b']['y'].mean())
print('La media del número de moléculas en c es:', df_a[df_a['z'] == 'c']['y'].mean())
```

```python
print('La mediana del número de moléculas en a es:', df_a[df_a['z'] == 'a']['y'].median())
print('La mediana del número de moléculas en b es:', df_a[df_a['z'] == 'b']['y'].median())
print('La mediana del número de moléculas en c es:', df_a[df_a['z'] == 'c']['y'].median())
```

La media y la mediana en el eje "y" tampoco es exacta ya que se estan considerando los datos completados con 0.

```python
round(df_a.describe(),2)
```

La gráfica muestra que a cierta temperatura se indica la cantidad de móleculas que se mueven a determinada velocidad. El punto más alto de cada curva indica la velocidad del mayor número de moléculas, y esta aumenta cuando la temperatura es mayor. Además,
medida que aumenta la temperatura, un mayor número de moléculas se mueven a mayor velocidad.
